## Scatter plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import random
```

```python
numdots = 20
# generate random x and y values between 0-9
xval = [x for x in range(numdots)]
yval = [int(random.random()*100) for _ in range(numdots)]
print('x=',xval)
print('y=',yval)

# draw a scatter figure for (x,y) values
plt.scatter(xval, yval, color='blue')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

plt.show()
```

```python
numdots = 100

# plot red dots

# random numbers list
xlist = [random.random() for _ in range(numdots)]
# another random numbers list
ylist = [ random.random() for _ in range(numdots)]
plt.scatter(xlist, ylist, color='red')

# plot blue dots

# random numbers list
xlist = [random.random() for _ in range(numdots)]
# another random numbers list
ylist = [ random.random() for _ in range(numdots)]
plt.scatter(xlist, ylist, color='blue')

# set legends and title
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.title('Random Data',fontsize=20)

plt.show()
```

### Let's use Titanic data to show the relationship between passengers' ages, fare paid, and survival

```python
# open titanic.csv file and read headers
fhand = open('titanic.csv')
headers = fhand.readline().strip()
headers
```

```python
# read all data, make lists of (1) ages, (2) fare paid, and (3) survival of passengers
age = [] 
fare = []
survival = []
for line in fhand:
    data = line.strip().split(',')
    age.append(float(data[4]))
    fare.append(float(data[-1]))
    survival.append('blue' if data[0]=='1' else 'red')
```

```python
# draw scatter plot of these three lists (ages by fare paid)
plt.figure(figsize=(10, 10), dpi=80)
plt.scatter(age, fare, color=survival, marker='o')
plt.title('Fare paid vs Age', fontsize=20, color='blue')
plt.xlabel('Age', fontsize=20, color='blue')
plt.ylabel('Fare', fontsize=20, color='blue')
plt.show()
```

## Line plots

```python
xval = [x for x in range(100)]
yval = [x**2 for x in xval]
yval2 = [10000 - x**2 for x in xval]
```

```python
plt.plot(xval,yval)
plt.plot(xval,yval2)
plt.show()
```

```python
xval = [x for x in range(-100,100,10)]
yval = [x**3- 2*x**2 + 500 for x in xval]
plt.plot(xval,yval,color='purple',marker='+')
plt.show()
```

## Bar graphs

```python
# Given the following dictionary with planets and their radiuses
planets = { 'mercury':1516, 'earth':3959, 'saturn':36184, 'mars':2106, 'neptune':15299, 'jupiter':43441, 'venus':3760, 'uranus':15759}

plt.figure(figsize=(10, 5), dpi=80)

# Plot planets and their radius data using a bar graph
plt.bar(planets.keys(),planets.values())
plt.title('Planets',fontsize=20)
plt.ylabel('radius (mi)', fontsize=15, color='blue')
plt.show()
```

```python
# Plot planets and their radius data using a bar graph horizontally
names = [x for x in planets.keys()]
radius= [y for y in planets.values()]
plt.figure(figsize=(10, 5), dpi=80)
plt.barh(names,radius,color='green')
plt.title('Planets',fontsize=20)
plt.xlabel('radius (mi)', fontsize=15, color='blue')
plt.show()
```
